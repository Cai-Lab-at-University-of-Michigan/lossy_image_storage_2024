```python
import glob
import numpy as np
from metrics import eval_metric_instance, eval_metric_instance_reverse
import pandas as pd
from tqdm.notebook import tqdm
```

```python
data_path = '/data/duanb/EM/CREMI_A/'
```

```python
models = ['U_Net', 'UNetplus', 'TransUnet']
```

```python
for gt in ['gt', 'uncompressed']:
    data = pd.DataFrame({
        'Compression Method': [],
        'Compressed Ratio': [],
        'Model': [],
        'FG_Dice': [],
        'ARAND' : [],
        'VOI_Merge' : [],
        'VOI_Split' : [],
        })
    for m in models:
        if gt == 'gt':
            gt_tif = f'{data_path}GT.tif'
        else:
            gt_tif = f'{data_path}uncompressed_{m}.tif'
            
        for tif in tqdm(sorted(glob.glob(data_path + f'*_{m}.tif'))):
            if tif == gt_tif:
                continue
            else:
                if 'uncompressed' in tif:
                    cmethod = 'uncompressed'
                    cratio = 1.0
                else:
                    cmethod = tif.split('/')[-1].split('_cs')[0].split('EM_')[-1]
                    cratio = float(tif.split('/')[-1].split('_cs')[-1].split('_')[0])
                fg_dice, arand, voi_over, voi_under = eval_metric_instance_reverse(tif, gt_tif)
                data = pd.concat([data, pd.DataFrame([{
                    'Compression Method': cmethod,
                    'Compressed Ratio': cratio,
                    'Model': m,
                    'FG_Dice': fg_dice,
                    'ARAND' : arand,
                    'VOI_Merge' : voi_over,
                    'VOI_Split' : voi_under,
                }])], ignore_index=True)
    data.to_csv(f'EM_stat_{gt}.csv')
```
